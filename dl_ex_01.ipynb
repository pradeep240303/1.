```python
#rank 0 tensor (scalars)
import numpy as np
rank0 = np.array(2)
print(rank0,rank0.shape,rank0.ndim,rank0.dtype)
```

```python
#rank 1 tensor(vectors)
rank1 = np.array([1,2,3])
print(rank1,rank1.shape,rank1.ndim,rank1.dtype)
```

```python
#rank 2 tensor(vectors)
rank2 = np.array([[1,2,3],[4,5,6]])
print(rank2,rank2.shape,rank2.ndim,rank2.dtype)
```

```python
#access specific element from 2d matrix
print(rank2[1][2])
```

```python
#rank 3 tensor (3D matrix)
rank3 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(rank3,rank3.shape,rank3.ndim,rank3.dtype)
```

```python
print(rank3[0][0][1])
print(rank3[0][1][1])
print(rank3[0][1][1])
```

```python
'''Applying operations 
      ones, zeros, fill, eye, identity '''
#ones with default float representation
ones = np.ones([4,5])
print(ones)

#ones with integer representation
ones = np.ones([4,5], dtype = int)
print(ones)

#zeros
zeros = np.zeros([2,3,3])
print(zeros)
```

```python
full1= np.full([5,4],2)
print(full1)
```

```python
eyes=np.eye(4,4)
print(eyes)
```

```python
iden=np.identity(4)
print(iden)
```

```python
import tensorflow as tf 
#scalar 
scalar = tf.constant(3)
print(scalar)
scalar = tf.Variable(3)
print(scalar)

#vector
vector = tf.constant([1,2])
print(vector)

#2D matrix
matrix2d = tf.constant([[1,2],[4,5]])
print(matrix2d)

#ones
ones = tf.ones([3,2])
print(ones)
#zeros
zeros = tf.zeros([4,4])
print(zeros)
#fill
fill = tf.fill([4,4],4)
print(fill)
```
